```python
# Write a function that calculates the (Hamming) distance between two words of equal length.
def hamming_distance(str1, str2):
    if not len(str1) == len(str2):
        print("error")
        return -1
    else: 
        distance = 0 
        for i in range(len(str1)): 
            if not str1[i] == str2[i]: 
                distance = distance + 1
    return distance
```
